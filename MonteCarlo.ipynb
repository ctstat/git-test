1. when # of estimates increase --> the mean of estimates closes to the true value

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = 1.5 
size1 = 30                   # num of x_hat
size2 = 30000


np.random.seed(size1)
errors1 = np.random.standard_normal(size = size1)
xhats1 = x + errors1

np.random.seed(size2)
errors2 = np.random.standard_normal(size = size2)
xhats2 = x + errors2
```

```python
# making subplots objects
fig, ax = plt.subplots(1, 2)
ax[0].hist(xhats1, color = 'blue', label=f"{size1}")
ax[1].hist(xhats2, color = 'red', label=f"{size2}")
plt.legend(loc='upper right')
plt.show()
```

```python
print(f"mean estimates of {size1}: {round(np.mean(xhats1),4)}")
print(f"mean estimates of {size2}: {round(np.mean(xhats2),4)}")
```
